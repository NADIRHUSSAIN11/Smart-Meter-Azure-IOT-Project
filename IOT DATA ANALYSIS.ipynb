```python
df=spark.read.csv('dbfs:/mnt/iotdata/*/*/*/*/*.csv',inferSchema=True,header=True)
```

```python
display(df)
```

```python
df.write.mode('overwrite').saveAsTable('meterdata')
```

```python

avg_temp=spark.sql('SELECT id,Count(temp) as count ,avg(temp) as avg_temp FROM meterdata GROUP BY id ORDER BY id')
```

```python
avg_temp.write.mode('overwrite').saveAsTable('avgtemp')
```

```python
avg_voltage=spark.sql('SELECT id,Count(voltage) as count,avg(voltage) as avg_voltage FROM meterdata GROUP BY id ORDER BY id')
```

```python
avg_voltage.write.mode('overwrite').saveAsTable('avgvoltage')
```
